```python
from PIL import Image
import os
import mnist
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

```python
path='digits'
digit_files=os.listdir(path)


digit_files=['digits/'+i for i in digit_files]

digits_list=[i[7:-5] for i in digit_files]

print(digit_files,'\n')
print(digits_list)
```

```python
x_train=mnist.train_images()
y_train=mnist.train_labels()
```

```python
print("Datatype of x_train: {}".format(type(x_train)))
print("Datatype of y_train: {}".format(type(y_train)))
```

```python
print("Shape of x_train: {}".format(x_train.shape))
print("Shape of y_train: {}".format(y_train.shape))
```

```python
print("Number of dimensions in x_train: {}".format(x_train.ndim))
print("Number of dimensions in y_train: {}".format(y_train.ndim))
```

```python
x_train=x_train.reshape((-1,28*28))
```

```python
print("Shape of x_train: {}".format(x_train.shape))
```

```python
print(x_train[0])
```

```python
x_train=x_train/256
```

```python
print(x_train[0])
```

```python
clf=MLPClassifier(solver='adam',activation='relu',hidden_layer_sizes=(64,64))
```

```python
clf.fit(x_train,y_train)
```

```python
def accuracy(cm):
    diagonal=cm.trace()
    elements=cm.sum()
    return diagonal/elements
```

```python
# for path in digit_files:
#     print(path)
data=[]
for path in digit_files:
    img=Image.open(path)
    data.append(list(img.getdata()))

for i in data:
    print(i,"\n\n")
```

```python
for i in data:
    for j in range(len(i)):
        i[j]=255-i[j]

for i in data:
    print(i,"\n\n")
```

```python
for i in range(len(data)):
    num=data[i]
    num=np.array(num)/256
#     print(num,'\n\n')
    id=clf.predict([num])
    
    print(id,'\t',digit_files[i])
```
